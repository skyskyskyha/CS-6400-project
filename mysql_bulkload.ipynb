```python
import json
```

```python



# business
input_file = "reduced_businesses.json"
output_file = "reduced_businesses.dat"

with open(input_file, "r", encoding="utf-8") as f:
    data = [json.loads(line) for line in f]


fields = [
    "business_id", "name", "address", "city", "state",
    "postal_code", "latitude", "longitude", "stars",
    "review_count", "is_open"
]

days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
default_hours = "11:0-21:0"

with open(output_file, "w", encoding="utf-8") as f:
    for item in data:
        if item is None:
            continue
        # Process fields
        row = "|".join([str(item.get(field, "")).replace('', '').replace('|', ' ') for field in fields])

        # Process hours
        hours = item.get("hours", {}) or {}
        print(hours)
        hours_row = "|".join([str(hours.get(day, default_hours)).replace('|', ' ') for day in days])

        # Combine fields and hours
        final_row = row + "|" + hours_row
        f.write(final_row + "\n")


```

```python
# review
review_input_file = "yelp_academic_dataset_review.json"
review_output_file = "yelp_academic_dataset_review.dat"
with open(review_input_file, "r", encoding="utf-8") as f:
    review_data = [json.loads(line) for line in f]

review_fields = [
    "review_id", "user_id", "business_id", "stars",
    "date",
    # "text",
    "useful", "funny", "cool"
]

with open(review_output_file, "w", encoding="utf-8") as f:
    for item in review_data:
        row = "|".join([str(item.get(field, "")).replace('|', ' ') for field in review_fields])
        f.write(row + "\n")

```

```python
input_file = "reduced_tips.json"
output_file = "reduced_tips.dat"

fields = ["user_id", "business_id", "text", "date", "compliment_count"]

with open(input_file, "r", encoding="utf-8") as f:
    data = [json.loads(line) for line in f]

with open(output_file, "w", encoding="utf-8") as f:
    for item in data:
        row = "|".join([str(item.get(field, "")).replace('|', ' ').replace('\n', ' ') for field in fields])
        f.write(row + "\n")
```

```python
# friends
import json


input_file = "reduced_users.json"
output_file = "reduced_friends.dat"

with open(input_file, "r", encoding="utf-8") as f:
    data = [json.loads(line) for line in f]

with open(output_file, "w", encoding="utf-8") as f:
    for item in data:
        user_id = item.get("user_id", "")
        friends = item.get("friends", [])
        
        if not friends:
            continue
        
        for friend_id in friends:
            f.write(f"{user_id}|{friend_id}\n")
 
```

```python
input_file = "reduced_businesses.json"
output_file = "reduced_categories.dat"

with open(input_file, "r", encoding="utf-8") as f:
    data = [json.loads(line) for line in f]

with open(output_file, "w", encoding="utf-8") as f:
    for item in data:
        business_id = item.get("business_id", "")
        categories = item.get("categories", "")
        
        if categories:
            category_list = [cat.strip() for cat in categories.split(",")]
            
            for category in category_list:
                f.write(f"{business_id}|{category}\n")
```
